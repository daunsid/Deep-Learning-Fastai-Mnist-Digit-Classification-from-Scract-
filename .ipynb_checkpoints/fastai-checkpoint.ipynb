```python
speed = torch.randn(20)*3 + 0.75*(time-9.5)**2 + 1
plt.scatter(time, speed);
```

```python
def f(t, params):
    a, b, c = params
    return a*(t**2) + (b*t) + c
```

```python
def mse(preds, targets):
    return ((preds-targets)**2).mean()
```

```python
params = torch.randn(3).requires_grad_()
```

```python
orig_params = params.clone()
orig_params
```

```python
def show_preds(preds, ax=None):
    if ax is None: ax=plt.subplots()[1]
    ax.scatter(time, speed)
    ax.scatter(time, to_np(preds), color='red')
    ax.set_ylim(-300, 100)
```

```python
lr = 1e-5
params.data = params.data - lr *params.grad.data
params.grad = None
```

```python
preds = f(time, params)
mse(preds, speed)
```

```python
show_preds(preds)
```

```python
def apply_step(params, prn=True):
    preds = f(time, params)
    loss = mse(preds, speed)
    loss.backward()
    params.data -= params.grad.data
    params.grad = None
    if prn:
        print(loss.item())
    return preds
```

```python
def calc_grad(xb, yb, model):
    preds = model(xb)
    loss = mnist_loss(preds, yb)
    loss.backward()
```

```python
def train_epoch(model, lr, params):
    for xb, yb, in dl:
        calc_grad(xb, yb, model)
        for p in params:
            p.data -= p.grad*lr
            p.grad.zero_()
```

```python
def batch_accuracies(xb, yb):
    preds = xb.sigmoid()
    correct = (preds>0.5).float() == yb
    return correct.float().mean
```

```python
class Linear(nn.Linear):
    def __init__(self, input_shape, out_shape):
        super(Linear, self).__init__()
        self.input_shape = input_shape
        self.out_shape = out_shape
        
        
    def __sizeof__(self):
        return tensor(self.input_shape).shape, tensor(self.out_shape).shape
    
    
```

```python
class BasicOptim:
    def __init__(self, params, lr): self.params,self.lr= list(params), lr
        
    def step(self, *args, **kwargs):
        for p in self.params: p.data -= p.grad.data * self.lr
    
    def zero_grad(self, *args, **kwargs):
        for p in self.params: p.grad = None
```
